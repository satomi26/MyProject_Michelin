Data cleanings that I want to do:
1. Price. I want to create new column for price. for example $$$$ -= 4, $$$ = 3, YY = 2

```python
# 2. split location into city and country
df[["City", "Country"]] = df.Location.str.split(",", expand = True )
df.head()
```

there are spaces in front of countrys name. remove them with lstrip().  and czechia and czech republic are same. So I will replace czechia to czech republich.

```python
df["Country"] = df["Country"].str.lstrip()
```

```python
df["Country"] = df["Country"].replace('Czechia','Czech Republic')
```

```python
# I wanted to convert from country to country code. Then came problem, that china mainland is unknown. so I will replace value from china mainland to china.
df['Country'] = df['Country'].replace(['China Mainland'], 'China')
```

```python
# Now i have another problem KeyError: "Invalid Country Name: 'Taipei & Taichung'"
df['Country'] = df['Country'].replace(['Taipei & Taichung'], 'China')

```

Now I know which rows have None value in Country Column. I try to fill the country name into country column.


```python
# first i try to do for singapore
df.loc[df.Location == 'Singapore', 'Country'] = 'Singapore'
```

```python
df['Country'][df.Location == 'Singapore'] = 'Singapore'
```

```python
nan_in_country
```

```python
df["Country"].unique()
# now singapore is created.
```

```python
# I will do this with other Locations that have none country column, 'Macau', 'Hong Kong', 'Dubai', 'Luxembourg','Abu Dhabi'. I will make it hongkong, macau to china. only for the purpose to give it country code to make continent columns
df['Country'][df.Location == 'Macau'] = 'China'
df['Country'][df.Location == 'Hong Kong'] = 'China'
df['Country'][df.Location == 'Dubai'] = 'United Arab Emirates'
df['Country'][df.Location == 'Luxembourg'] = 'Luxembourg'
df['Country'][df.Location == 'Abu Dhabi'] = 'United Arab Emirates'
```

```python
df['Country'] = df['Country'].replace(['Türkiye'], 'Turkey')
```

```python
df['Country'] = df['Country'].replace(['Hong Kong SAR China'], 'China')
```

```python
df["Country"].unique()
```

```python
import pycountry_convert as pc
```

```python
f = lambda x: pc.country_name_to_country_alpha2(x, cn_name_format="default")
df['CountryCode'] = df['Country'].apply(f)
```

```python
df.head()
```

```python
df["CountryCode"].unique()
```

```python
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
```

```python

f = lambda x: pc.country_alpha2_to_continent_code(x)
df['Continent'] = df['CountryCode'].apply(f)
```

```python
df["Continent"].unique()
```

```python
f = lambda x: pc.convert_continent_code_to_continent_name(x)
df['ContinentName'] = df['Continent'].apply(f)
```
